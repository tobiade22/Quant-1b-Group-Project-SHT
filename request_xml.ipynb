```python
import requests

# URL of the XML resource
url = 'https://www.theyworkforyou.com/pwdata/scrapedxml/regmem/regmem2023-10-30.xml'

# Fetch the XML data as a string
response = requests.get(url)
xml_string = response.text  # Get the content as a string

```

```python
import xml.etree.ElementTree as ET

# Parse the XML string into an ElementTree object
root = ET.fromstring(xml_string)
```

```python
for regmem in root.findall('regmem'):
    name = regmem.get('membername')
    print(f"Member Name: {name}")
```

```python
person_id = "uk.org.publicwhip/person/25661"  # Example ID for Bim Afolami
for regmem in root.findall(f".//regmem[@personid='{person_id}']"):
    employment_earnings = regmem.findall(".//category[@name='Employment and earnings']/record/item")
    print(f"Employment and Earnings Entries for {regmem.get('membername')}: {len(employment_earnings)}")
```
